```python
import os
path = os.path.expanduser('~/GIT/Research_PatrickP/periodic_2d_disads/res/')
text_files = [f for f in os.listdir(path) if f.endswith('.txt')]
text_files
```

```python
path ='~/GIT/Research_PatrickP/periodic_2d_disads/res/'
# filenames = glob.glob(path + "/*.txt")
# creating empty list
dataFrames = list()
# iterating through CSV file in current directory
for text_file in text_files:
    dataFrames.append(pd.read_csv(path+text_file, header=0, sep=' '))
    print(text_file)
```

```python

# Concatenate all data into one DataFrame
merged_frame = pd.concat(dataFrames,axis=1)
# print merged dataframe
print(merged_frame)
```
